## Elimino Caracteres especiales

```python
special_characters = (",",".",'"',":","-")
for char in special_characters:
    readme = readme.replace(char," ")
print (readme)
```

## Resuelvo Ejercicio 2 sin Counter

```python
words = readme.split()
word_frecuence ={}

for word in words:
    if len(word) > 4:
        if word.lower() in word_frecuence:
            word_frecuence[word.lower()]+=1
        else:    
            word_frecuence[word.lower()] = 1

word_frecuence
```

## Resuelvo Ejercicio 2 con Counter

```python
from collections import Counter

cnt = Counter()
words = readme.lower().split()

for word in words:
   if len(word) >4:
      cnt[word] +=1

cnt.most_common(1)

```
